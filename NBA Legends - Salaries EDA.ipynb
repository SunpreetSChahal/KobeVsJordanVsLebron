### Salaries

Let's begin with something simple - their salaries. This won't provide any insight into who's the best, but there's no harm to look into it.

```python
import pandas as pd
import numpy as np 
import datetime as dt
import matplotlib.pyplot as mpy
import matplotlib as mpt
import seaborn as sns
```

```python
salaries = pd.read_csv("D:\PROJECTS\Kobe-V-Lebron-V-Jordan âœ”\salaries.csv")
```

```python
salaries.columns
```

#### Cleaning & Manipulation

```python
salaries.isnull().sum()
```

```python
kb_salary = salaries.loc[salaries['Player'] == 'Kobe Bryant']
mj_salary = salaries.loc[salaries['Player'] == 'Michael Jordan']
lj_salary = salaries.loc[salaries['Player'] == 'Lebron James']
```

```python
## Split up the Season years and Salaries per player
kb_s_season = kb_salary['Season']
kb_s_salary = kb_salary['Salary']

mj_s_season = mj_salary['Season']
mj_s_salary = mj_salary['Salary']

lj_s_season = lj_salary['Season']
lj_s_salary = lj_salary['Salary']
```

#### Visualisation

```python
mpy.figure(figsize=(25,15))
mpy.title('Salaries')

mpy.subplot(331)
mpy.bar(kb_s_season, kb_s_salary, label= 'K. Bryant', color= 'blue')
mpy.xlabel('Season')
mpy.xticks(rotation= 'vertical')
mpy.ylabel('Salary')
mpy.legend()

mpy.subplot(334)
mpy.bar(mj_s_season, mj_s_salary, label= 'M. Jordan', color= 'green')
mpy.xlabel('Season')
mpy.xticks(rotation= 'vertical')
mpy.ylabel('Salary')
mpy.legend()

mpy.subplot(337)
mpy.bar(lj_s_season, lj_s_salary, label= 'L. James', color= 'red')
mpy.xlabel('Season')
mpy.xticks(rotation= 'vertical')
mpy.minorticks_on
mpy.ylabel('Salary')
mpy.legend()
```

Though all players were paid different amounts, the general trend of the amount they recieved was positive.
 
As each player dominated their rivals and showed great talent, their fan pools grew. The more money they brought in to the NBA, the more they recieved.
