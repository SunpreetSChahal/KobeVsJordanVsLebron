Kobe Bryant, Michael Jordan and Lebron James, all have high-quality and valuable careers All being awarded many achievements throughout their careers. 

However, after analysing their careers we can conclude the following:

#### Kobe Bryant -

- Appears to the least consistent out of the three.
- Most volitile career
- Was the best performer from the three; 
    - Scored the highest points in a game.
    - Most field goals in a game.
    - Highest 3 pointers in a game.

#### Michael Jordan - 

- Has had a consistently higher perfoming career:
    - Highest average points scored.
    - Highest average free throws scored.
- The context behind a drop in Jordan's performance or missing data, is due to his 2 retirements;
    - October 1993 - 1994 (1st retirement)
    - January 1999 - 2001 (2nd retirement)
    
#### Lebron James - 

- Has the most consistent career from the three legends.
- We haven't seen a recent drop in performance in his career - other than blocks.
- Due to his career being the most consistent and his statistics rising in assists, points, field goals and free throws - this idicates he isn't slowing show. 
- Several years of strong performance is still possible. 

#### Who is the G.O.A.T? 

We won't truly know who was the *greatest of all time* until the sport of basketball ends.

However there are a few perspectives to this question:

- Kobe Bryant could be consisdered the G.O.A.T due to him being the strongest performer from the three, however, his career was also the most volitile. 
- If this question was looking for the player who was the highest-performer, Kobe Bryant takes that title.


- Micheal Jordan could be consisdered the G.O.A.T due to him having a more consistent, higher-performing career. 
- However, nearing the end of his career we can see a significant drop in performance, notably, a drop in the points he scored after his second retirement (or break).
- If the G.O.A.T title was to be given to the player with notable career, them Michael Jordan would take that title. Featuring in movies such as Space Jam and having the most popular shoe line from the others, would make MJ a strong candidate for the G.O.A.T title. However, I see this title fitting our final candidate the best.


- Lebron James could be consisdered the G.O.A.T due to having the most consistent career from the three. He has been peforming at a high-level throughout his time at the NBA. 
   - Due to not slowing down as of recent, we can anticipate several more years of strong, consistent performance.
- If this question was looking for the player who performed at a high-level, consistently. Lebron James takes that title, and personally I would agree with this, he's the most consistent out of the three players and shows no sign of slowing down.
